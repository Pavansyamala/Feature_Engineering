```python
## Linear Interpolation
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x_new = np.linspace(x.min() , x.max() , 2*x.shape[0])
```

```python
## We need to find the y_new corrosponding to the x_new as it is linear study
```

```python
y_new = np.interp(x_new , x , y) ## interp is the function available in the numpy for 1d data
```

```python
plt.scatter(x_new,y_new)
plt.show()
```

```python
x1 = np.array([0,30,45,60,90,120,145,160,170,180,200,270,300,360])
d = np.deg2rad(x1)
y1 = np.sin(d)
```

```python
plt.plot(x1,y1)
plt.show()
```

```python
x1_new = np.linspace(x1.max() , x1.min() , 2*len(x1))
```

```python
d = np.deg2rad(x1_new)
```

```python
y1_new = np.interp(x1_new,x1,y1)
```

```python
plt.plot(x1_new,y1_new)
plt.show()
```

```python
## Using Scipy
```

```python
from scipy.interpolate import interp1d
```

```python
x2 = np.array([1,2,3,4,5])
y2= np.array([1,4,9,16,25])
```

```python
x2_new = np.linspace(x2.min(),x2.max() , 2*len(x2))
```

```python
func = interp1d(x2,y2,kind = 'cubic')
```

```python
y2_new = func(x2_new)
```

```python
plt.plot(x2_new,y2_new)
plt.show()
```

```python
## Polynomial Interpolation
```

```python
x3 = np.array([1,2,3,4,5,6,7,8])
y3 = x3**5
```

```python
x3_new = np.linspace(x3.min(),x3.max(),2*len(x3))
```

```python
dfit_3 = np.polyfit(x3,y3,5)
```

```python
y3_new = np.polyval(dfit_3 , x3_new)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.plot(x3,y3)
plt.title('Original')
plt.subplot(1,2,2)
plt.plot(x3_new,y3_new)
plt.title('Required')
plt.show()
```
