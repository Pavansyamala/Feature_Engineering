```python
from sklearn.datasets import make_classification
```

```python
data , labels = make_classification(n_samples = 1000 ,n_features = 2 , n_redundant=0 , weights=[0.90 , 0.1], n_informative=1 , n_clusters_per_class=1)
```

```python
df = pd.DataFrame({'Feature 1' : data[:,0], 'Feature 2':data[:,1]})
```

```python
df['Label'] = labels
```

```python
df['Label'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.relplot(data = df , x = 'Feature 1' , y = 'Feature 2' , hue = 'Label')
plt.show()
```

```python
import imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
over_sample = SMOTE()
```

```python
x,y = over_sample.fit_resample(df.drop(columns = ['Label']).values, df['Label'])
```

```python
data_df = pd.DataFrame(x , columns = ['f1' , 'f2'])
```

```python
data_df['Label'] = y
```

```python
data_df.shape
```

```python
sns.relplot(data = data_df , x = 'f1' , y = 'f2' , hue = 'Label')
plt.show()
```
